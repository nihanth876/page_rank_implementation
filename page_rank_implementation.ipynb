```python
def manhattan_distance(a, b):
    return np.abs(a - b).sum()

def simple_page_rank_algorithm(df,threshold,weighted=False):
    probs = {}
    for i in df.columns:
        if i=='page_name':
            continue
        else:
            probs[i]= 1/sum(df[i])

    probability_matrix = pd.DataFrame()
    probability_matrix['page_name'] = df['page_name']
    for i in df.columns:
        if i=='page_name':
            continue
        probability_matrix[i] = df[i]*probs[i]
    
    r = pd.DataFrame()
    if weighted:
        temp = list(df.columns)
        temp.remove('page_name')
        x = df[temp].sum().sum()
        df11 = pd.DataFrame(df[temp].T.sum())
        df12 = df11.T
        df12.columns = temp
        df12 = df12/x
        r = df12
        
    else:
        prob = pd.DataFrame([1/(len(df.columns)-1)])
        print(prob)
        for i in df.columns:
            if i=='page_name':
                continue
            r[i] = prob[0]
    print(r)    
    metric = float('inf')
    #threshold = 0.005
    col_names = r.columns
    while metric>threshold:
        #print(r)
        temp = list(df.columns)
        temp.remove('page_name')
        temp_r = probability_matrix[temp].dot(r.T).T

        a = np.array(r)
        b = np.array(temp_r)
        r = temp_r
        r.columns = col_names
        metric = manhattan_distance(a, b)
        print(metric)
    return r
    
```
